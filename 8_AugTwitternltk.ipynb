```
tweet_text = []
for line in open('endgametweets1.txt', 'r'):
    tweet_text.append(line.strip())
```

```
tweet_text
```

```
def convert(s): 
  
    # initialization of string to "" 
    new = "" 
  
    # traverse in the string  
    for x in s: 
        new += x  
  
    # return string  
    return new 
```

```
convert(tweet_text)
```

```
text=convert(tweet_text)
```

```
print(sent_tokenize(text))
```

```
print(word_tokenize(text))
```

```
import string
```

```
text2=text.translate(str.maketrans('', '', string.punctuation))
```

```
text2
```

```
tokenizedtext=word_tokenize(text2)
```

```
print(tokenizedtext)
```

```
from nltk.stem import PorterStemmer
```

```
ps=PorterStemmer()
```

```
for w in tokenizedtext:
  print(w," : ",ps.stem(w))
```

```
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer

vec = CountVectorizer()
X = vec.fit_transform(tokenizedtext)
df = pd.DataFrame(X.toarray(), columns=vec.get_feature_names())
print(df)
```
